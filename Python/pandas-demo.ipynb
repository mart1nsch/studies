Criar uma coluna nova

```python
df['ENDERECO_COMPLETO'] = df['ENDERECO'] + ", " + df['NUMERO'].astype(str) + ", " + df['BAIRRO']
df
```

Remover coluna criada

```python
df = df.drop('ENDERECO_COMPLETO', axis=1)
df
```
